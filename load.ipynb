```python
import pandas as pd

from IPython.display import display
```

```python
# Load the first two files (uploaded ones)
file1 = 'Hourly-load-by-area-and-region-2011-to-2017-.xlsx'
file2 = 'Hourly-load-by-area-and-region-2017-2020.xlsx'
file3 = 'Hourly-load-by-area-and-region-May-2020-to-Oct-2023.xlsx'
file4 = 'Hourly-load-by-area-and-region-Nov-2023-to-Dec-2024.xlsx'
```

```python
# Load the specific sheets
df1 = pd.read_excel(file1, sheet_name='Load by AESO Planning Area')
df2 = pd.read_excel(file2, sheet_name='Load by Area and Region')
df3 = pd.read_excel(file3, sheet_name='Sheet1')
df4 = pd.read_excel(file4, sheet_name='Sheet1')

# Display first few rows of same type datasets
print("Preview of df1:")
display(df1.head())

print("\nPreview of df2:")
display(df2.head())
```

```python
# Display first few rows of same type datasets
print("Preview of df1:")
display(df3.head())

print("\nPreview of df2:")
display(df4.head())
```

```python
#The last 5 rows should be after merging
display(df4.tail())
```

```python
# Convert 'Date' column to datetime format
df1['Date'] = pd.to_datetime(df1['DATE'], errors='coerce')
df2['Date'] = pd.to_datetime(df2['DATE'], errors='coerce')

# Convert 'Hour Ending' to integer (handling non-numeric cases if any)
df1['Hour Ending'] = df1['HOUR ENDING'].astype(str).str.extract('(\d+)')[0].astype(float) - 1
df2['Hour Ending'] = df2['HOUR ENDING'].astype(str).str.extract('(\d+)')[0].astype(float) - 1

# Create 'DT_MST' column
df1['DT_MST'] = df1['Date'] + pd.to_timedelta(df1['Hour Ending'], unit='h')
df2['DT_MST'] = df2['Date'] + pd.to_timedelta(df2['Hour Ending'], unit='h')

# Drop old 'Date', 'Hour Ending', 'DATE', and 'HOUR ENDING' columns
df1.drop(columns=['Date', 'Hour Ending', 'DATE', 'HOUR ENDING'], inplace=True)
df2.drop(columns=['Date', 'Hour Ending', 'DATE', 'HOUR ENDING'], inplace=True)

# Move 'DT_MST' column to the front
df1 = df1[['DT_MST'] + [col for col in df1.columns if col != 'DT_MST']]
df2 = df2[['DT_MST'] + [col for col in df2.columns if col != 'DT_MST']]

# Display the first few rows
display(df1.head())
display(df2.head())
```

```python
# Load your DataFrames (assuming they're already defined)
merged_df = pd.concat([df1, df2, df3, df4], ignore_index=True)

# Display the first few rows to verify
display(merged_df.head())
display(merged_df.tail())
```

```python
# Save the merged DataFrame to an Excel file
file_path = "merged_data.xlsx"
merged_df.to_excel(file_path, index=False)

print(f"File saved as {file_path}. You can download it now!")
```
